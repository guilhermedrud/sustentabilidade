```python
dfs = {}
for ano in anos:
    dfs_ano = []
    for folder in glob.glob(ano+"/*"):
        #print(folder)
        dimension_df = []
        for file in glob.glob(folder+"/*"):
            dimension_df.append(pd.read_csv(file,header=[0,1],index_col=0))
        dimension_df = pd.concat(dimension_df,axis=0,join="outer",ignore_index=False,keys=None,levels=None,names=None,verify_integrity=False)
        dfs_ano.append(dimension_df)
    dfs[ano] = dfs_ano
```

```python
def weighted_average(df,data_col,weight_col,by_col):
    df['_data_times_weight'] = df[data_col]*df[weight_col]
    df['_weight_where_notnull'] = df[weight_col]*pd.notnull(df[data_col])
    g = df.groupby(by_col)
    result = g['_data_times_weight'].sum() / g['_weight_where_notnull'].sum()
    del df['_data_times_weight'], df['_weight_where_notnull']
    return result.to_frame(name='value')

def corrige(string):
    return string.replace(",",'.')

def strip(string):
    return string.strip()

def get_labels(ano,df, label_df, cont_dict, ac_dict):
    #df.index = df.index.apply(str.strip)
#     print(cont_dict['COMPANHIA ENERGETICA DE PERNAMBUCO'])
    empresas = list(map(cont_dict.get,df.index.to_list()))
    #print(empresas)
    #print(df.index.to_list())
    labels = []
    #print(empresas)
#     for a in (df.index.to_list()):
#         print(a)
#         print(cont_dict[a])
    
    for i in empresas:
#         print(i)
#         print(ac_dict[ano][i])
        labels.append(label_df.loc[ac_dict[ano][i]]['value'])
    return labels
```

```python
labeled_dfs = {}
for ano in anos:
    print(ano)
    all_empresas = pd.concat(dfs[ano],axis=1, join="outer").index.sort_values().unique().to_list()
    url2 = 'https://respostas.iseb3.com.br/cicles/'+ano
    ht = urllib.request.urlopen(url2).read()
    soup = BS(ht)
    lista = soup.find_all("td", {"class": ["tit_holding","tit_controlada"]})

    pairs = []
    for item in lista:
        classe = re.findall(r"\".*\"",str(item))[0][1:-1]
        #print(classe,item.text)
        pairs.append((item.text,classe))

    controladas = []
    controladas_dict = {}
    empresa_atual = ""
    for empresa,classe in pairs:
        empresa = empresa.strip()
        if classe == "tit_holding" and empresa not in list(controladas_dict.keys()):
            controladas_dict[empresa] = empresa
            empresa_atual = empresa
        else:
            controladas_dict[empresa] = empresa_atual
            controladas.append(empresa)
    empresas_mae = [empresa for empresa in all_empresas if empresa not in controladas]
    #print(controladas_dict)
    acao_df = pd.read_csv("../data/acoes/"+ano+".csv")
    if ano =='2020':
        #acao_df[ano] = acao_df[ano].apply(corrige).astype(float)
        controladas_dict['Sendas Distribuidora S.A'] = 'COMPANHIA BRASILEIRA DE DISTRIBUICAO'
        controladas_dict['TIM S.A']= 'TIM PARTICIPACOES S.A.'
        controladas_dict['PETROBRAS DISTRIBUIDORA S.A'] = 'PETROBRAS DISTRIBUIDORA S.A'
        acao_df['value'] = acao_df[ano]
        acao_df.index = acao_df.empresa
        #print(acao_df)
        concat_df = pd.concat(dfs[ano],axis=1, join="outer")
        concat_df = concat_df.groupby(concat_df.columns, axis=1).max()
        concat_df.index = concat_df.index.map(strip)
        #print(concat_df.index)
        concat_df['lucro'] = get_labels(ano,concat_df,acao_df,controladas_dict,dicio_acoes)
        labeled_dfs[ano] = concat_df
    else:
        if ano=='2019':
            controladas_dict['TIM S.A']='TIM PARTICIPACOES S.A.'
            controladas_dict['PETROBRAS DISTRIBUIDORA S.A']='PETROBRAS DISTRIBUIDORA S.A'
        acao_df['PART. %'] = acao_df['PART. %'].apply(corrige).astype(float)
        acao_df[ano] = acao_df[ano].apply(corrige).astype(float)
        acoes_avg = weighted_average(acao_df,ano,'PART. %','NOME')
        concat_df = pd.concat(dfs[ano],axis=1, join="outer")
        concat_df = concat_df.groupby(concat_df.columns, axis=1).max()
        concat_df.index = concat_df.index.map(strip)
        concat_df['lucro'] = get_labels(ano,concat_df,acoes_avg,controladas_dict,dicio_acoes)
        labeled_dfs[ano] = concat_df
```

```python
def clean(cols):
    clean_dic = {}
    for col in cols:
        ccol = "".join(col)
        ccol = re.sub('PERGUNTA [0-9\.]+','PERGUNTA x',ccol)
        clean_dic[col] = ccol
    return clean_dic
```

```python
clean_dfs = []
for df in labeled_dfs.values():
    cols = df.columns.to_list()
    rename_dict = clean(cols)
    clean_df = df.rename(mapper=rename_dict,axis=1)
    clean_dfs.append(clean_df)
```

```python
grouped_clean = pd.concat(clean_dfs,axis=1, join="outer")
grouped_clean = grouped_clean.groupby(grouped_clean.columns, axis=1).max()
#grouped_clean.fillna(0)
```
