```python
import pandas as pd
import glob
import numpy as np

import os
```

```python
anos = ['2015','2016','2017','2018','2019','2020']
```

```python
dfs = {}
for ano in anos:
    dfs_ano = []
    for folder in glob.glob(ano+"/*"):
        dimension_df = []
        for file in glob.glob(folder+"/*"):
            dimension_df.append(pd.read_csv(file,header=[0,1],index_col=0))
        dimension_df = pd.concat(dimension_df,axis=0,join="outer",ignore_index=False,keys=None,levels=None,names=None,verify_integrity=False)
        dfs_ano.append(dimension_df)
    dfs[ano] = dfs_ano
```
